Lecture 12: Self-Supervised Learning

large scale training problem -> we need a lot of labels manually

self-supervised learning

dataset (no labels) -> model (encoder) -> pretext objective
    - pretext objective : define a task based on the data itself
    - but we learn with supervised learning objectives like classification or regression
    - train encoder with large dataset (이건 그냥 unsupervised learning이라고 할 정도로 그냥 데이터 셋을 encoding하는데 의미가 더 있음)
    - decoder, classifier, regressor (pretext objective)와 붙여서 학습시키므로써 encoder를 leran representation하는 효과를 가져옴
    - pretext objective는 어떤 form of framework
    - label이 없으므로 loss계산은 없고 그냥 pseudo label을 만들어내기만 함


dataset small (labels) -> model (pretrained encoder) -> downstream objective
    - 실제 사용하는 모델로 학습된 encoder를 이식하여 더 학습시킴
