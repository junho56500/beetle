Objectives:
- Follow ball after hitting on tee shot
- Estimate the spot to be put on
- Estimate position based on field map
- Assist player to find the ball in quick

Background:
- Camera is fixed. installed behind player on tee shot
- It can be multiple cameras (depth estimation)
- High resolution (more than FHD 1080p, recommend 2K(1440p))
- Frame rate (more than 50fps)

Constraints
- Not moving camera
- Not other razor sensor
- Only installed on tee shot. Hard to chase from the 2nd shot

Challenges
- Tracking small pixel
- Tracking fast movement
- Labeling
- Roll over
