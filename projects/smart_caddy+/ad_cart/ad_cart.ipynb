Objectives:
- generate safe path to the target point generated from track ball application and next course after course done
- control cart following the path in safe speed
- safe driving for new player has not been new field
- follow track based on course map for safety but follow the ball for convenience of player

Background:
- inconvenient to find ball and drive to new way finding course to course
- unsafe driving for hill, hole or lateral slope and speedy on curve
- unsafe driving for dynamic objects like other player, worker and animal

Conditions:
- Prior course map (like satellite map or mapping manually)
- 2 cameras (front, rear) + extended
- E2E system from vision to control or prediction of path

Constraints
- adverse weather
- camera only

Challenges
- Hills
- hole
- embedded to control cart vehicle
